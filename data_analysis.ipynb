```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 读取数据集（假设已上传到 Colab 运行环境）
df = pd.read_csv('Business_Management_Dataset.csv')

# 选取指定字段
columns_to_plot = ['Employee_Satisfaction', 'Employee_Turnover', 'Market_Share']

# 创建子图
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, col in enumerate(columns_to_plot):
    # 使用 density=True 让纵轴显示概率密度，kde 显示核密度曲线
    sns.histplot(
        data=df,         # 明确传入数据
        x=col,           # 指定绘制的列
        ax=axes[i],      # 指定子图位置
        kde=True,        # 显示核密度估计曲线
        density=True     # 纵轴为概率密度
    )
    axes[i].set_title(f'{col} 直方图')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('概率密度')  # 调整纵轴标签

plt.tight_layout()
plt.show()
```
